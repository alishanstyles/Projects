```python
sns.histplot(data = df,
            x = 'claim_status')
plt.title('Distribution of Claim Status')
plt.xlabel('Claim Status')
plt.ylabel('Count');
```

> There appear to be more claim statuses compared to opinions. 

```python
sns.histplot(data = df,
             x ='video_id')
plt.title('Distribution of Video ID')
plt.xlabel('Video ID')
plt.ylabel('Count');
```

> Video with a id of 0.9  are more common.

```python
sns.histplot(data = df,
            x = 'video_duration_sec')
plt.title('Distribution of Duration of Video(Seconds)')
plt.xlabel('Video Duration (seconds)')
plt.ylabel('Count');
```

> It is common for videos to range from 15-45 seconds in length and the most common durations of time are 5, 25, 45 and 60 seconds with them all having a count of above 1000.

```python
plt.figure(figsize=(5,3))
sns.histplot(data = df,
              x = 'video_duration_sec', bins=range(0,61,5))
plt.title('Distribution of Duration of Video(Seconds)')
plt.xlabel('Video Duration (seconds)')
plt.ylabel('Count');
```

 > Videos with a duration of 60 seconds appear to be more common with a count of 2000.All videos are 5-60 seconds in length, and the distribution is uniform.

```python
plt.title('Video Duration (Secs)')
sns.boxplot(data = df,
           x = 'video_duration_sec');
```

```python
sns.histplot(data = df,
             x = 'video_view_count')
plt.title('Distribution of Video Views')
plt.xlabel('Video View Count')
plt.ylabel('Count');
```

> It is most common for videos to have 0.0 - 0.5 views with a count of almost 10000. 
This variable has a very uneven distribution, with more than half the videos receiving fewer than 100,000 views. Distribution of view counts > 100,000 views is uniform.

```python
sns.histplot(data = df,
             x = 'author_ban_status')
plt.title('Distribution of Author Ban Status')
plt.xlabel('Author Ban Status')
plt.ylabel('Count');
```

It appears there are more active authors with a count of well above 14000 compared to under review status with 2000 and banned status under 2000. 

```python
sns.histplot(data = df,
             x = 'verified_status')
plt.title('Distribution of Verified status')
plt.xlabel('Verified Status')
plt.ylabel('Count');
```

> According to the histogram there are more videos that are not verified than there are verified with there being more than 17500 unverified and less than 2500 verified.
