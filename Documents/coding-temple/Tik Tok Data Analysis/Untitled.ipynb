It appears there are more active authors with a count of well above 14000 compared to under review status with 2000 and banned status under 2000. 

```python
sns.histplot(data = df,
             x = 'verified_status')
plt.title('Distribution of Verified status')
plt.xlabel('Verified Status')
plt.ylabel('Count');
```

> According to the histogram there are more videos that are not verified than there are verified with there being more than 17500 unverified and less than 2500 verified. 

```python
plt.figure(figsize = (8,8))
sns.histplot(data = df,
             multiple = 'dodge',
             x = 'claim_status',
             hue = 'verified_status')
plt.title('Claims by verification status histogram')
plt.xlabel('Claim Status')
plt.ylabel('Count');
```

> It appears if a user is verified or unverified does not affect their claim status. The claim statuses appear to be equal for both verified and unverified. It does appear that unverified users are more likely to post cliams compared to verified users who are more likely to post opinions.

```python
plt.figure(figsize=(8,8))
sns.histplot(data = df,
            x = 'claim_status',
            hue = 'author_ban_status',
            multiple = 'dodge',
            shrink=0.9)
plt.title('Claim status by author ban status - counts')
plt.xlabel('Claim Status')
plt.ylabel('Count');
```

> Active authors tend to be more opinion based compared to banned authors who tend to be more cliam based. For both claims and opinions, there are many more active authors than banned authors or authors under review; however, the proportion of active authors is far greater for opinion videos than for claim videos. Again, it seems that authors who post claim videos are more likely to come under review and/or get banned.

```python
# Calculate the median views for all videos with an author ban status 
median = df.groupby('author_ban_status')['video_view_count'].median(numeric_only=True).reset_index()
```

```python
sns.barplot(data = median,
           x = 'author_ban_status',
           y = 'video_view_count')
plt.title('Video View Counts by Author Ban Status')
plt.xlabel('Author Ban Status')
plt.ylabel('Video View Count');
```

> The views for banned(non-active) authors appears to be higher than active authors. Based on this I would say view count is a good indicator of claim status.The median view counts for non-active authors are many times greater than the median view count for active authors. Since you know that non-active authors are more likely to post claims, and that videos by non-active authors get far more views on aggregate than videos by active authors, then video_view_count might be a good indicator of claim status.

```python
# Calculate the median view count for claim status.
df.groupby('claim_status')['video_view_count'].median()
```

```python
sns.barplot(data = df,
           x = 'claim_status',
           y = 'video_view_count')
plt.xlabel('Claim Status')
plt.ylabel('Video View Count')
plt.title('Claim Status by Video View Count');
```

```python
fig = plt.figure(figsize=(3,3))
plt.pie(df.groupby('claim_status')['video_view_count'].sum(), labels=['claim', 'opinion'])
plt.title('Total views by video claim status');
```

> The overall view count is dominated by claim videos even though there are roughly the same number of each video in the dataset.
